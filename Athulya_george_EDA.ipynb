```python
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
```

```python
region_distribution = customers['Region'].value_counts()
print(region_distribution)
region_distribution.plot(kind='bar', title="Customer Distribution by Region")
```

```python
signup_trends = customers['SignupDate'].dt.year.value_counts().sort_index()
signup_trends.plot(kind='line', title="Customer Signup Trends Over Time")
```

```python
category_distribution = products['Category'].value_counts()
category_distribution.plot(kind='pie', autopct='%1.1f%%', title="Product Distribution by Category")
```

```python

transactions['TotalValue'] = transactions['Quantity'] * transactions['Price']
total_revenue = transactions['TotalValue'].sum()
print(f"Total Revenue: ${total_revenue}")


top_products = transactions.groupby('ProductID')['TotalValue'].sum().sort_values(ascending=False)
print(top_products.head())
```
